```python
import requests
import re
import warnings
from bs4 import BeautifulSoup as bs
import pandas as pd

# 불필요한 경고 출력을 방지합니다.
warnings.filterwarnings('ignore')
```

```python
def gotowok(url) :
    html = requests.get(url, verify = False).text
    soup = bs(html,"lxml")
    menu_html = soup.select("div.menu_list > div.item > div.menu_info > p:nth-child(3)")

    menu_text = []
    for i in menu_html:
        text = i.get_text()
        menu_text.append(text)
    menu_merge = ",".join(menu_text)
    menu_merge_replace = menu_merge.replace(" ","").replace("\n","").replace("*메인메뉴는시기에따라변경될수있습니다.","")

    # 메뉴명
    menu_name = menu_merge_replace.split(",")

    # 가게이름 리스트 생성
    restaurant_name = []
    restaurant_category = []

    for num in range(len(menu_name)):
        restaurant_name.append("고투웍")
        restaurant_category.append("중식")
    
    gotowok_df = pd.DataFrame({"restaurant_name":restaurant_name,"restaurant_category":restaurant_category,"menu_name":menu_name})

    return(gotowok_df)
    
```
